```python
import tkinter as tk
from tkinter import messagebox
from spellchecker import SpellChecker

# Initialize the spell checker
spell = SpellChecker()

# Track incorrect words
incorrect_words = []

# Function to check words
def check_words():
    global incorrect_words

    user_input = entry.get()
    words = user_input.split()

    current_incorrect_words = [word for word in words if spell.correction(word) != word]

    if current_incorrect_words:
        for word in current_incorrect_words:
            if word not in incorrect_words:
                incorrect_words.append(word)
                
                # Display error message
                suggestions = ', '.join(spell.candidates(word))
                messagebox.showerror("Error", f"'{word}' is not available. Did you mean: {suggestions}?")
                
                if len(incorrect_words) == 2:
                    messagebox.showerror("Multiple Errors", f"Multiple incorrect words detected: {', '.join(incorrect_words)}.")
                    break
    else:
        messagebox.showinfo("Success", "No mistakes found! Your sentence is correct.")

# Main function to create and run the tkinter UI
def main():
    global entry

    # Create the main window
    root = tk.Tk()
    root.title("Sentence Checker")

    # Create the input field
    entry = tk.Entry(root, width=50)
    entry.pack(pady=20)

    # Create the check button
    check_button = tk.Button(root, text="Check Sentence", command=check_words)
    check_button.pack(pady=10)

    # Run the application
    root.mainloop()

# Ensure the UI opens when the script is run directly
if __name__ == "__main__":
    main()
```
